```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('vinayakphoto.jpeg')
img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_avgblur= cv2.blur(img_rgb,(5,5))
img_gauusian_blur=cv2.GaussianBlur(img_rgb,(5,5),0)
img_smooth=cv2.bilateralFilter(img_rgb,9,75,75)
l_title=["original","average blur","gaussian blur","smooth image"]
l_var=[img_rgb,img_avgblur,img_gauusian_blur,img_smooth]
for i in range (len(l_var)):
    plt.subplot(2,2,i+1);
    plt.title(l_title[i]);
    plt.imshow(l_var[i])
```
